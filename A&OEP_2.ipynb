```python
import numpy as np
from scipy import interpolate as itp
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
def ymod(x):
    return 10*np.sin(x)-2.5*np.cos(x)+x  #model
#PRVI SLUČAJ
x1=[0,2.5,5,7.5,10]
y1=[-2.5,10.488,-5.298,16.013,6.657]
plt.scatter(x1,y1,label='Slucaj 1')
plt.legend()
plt.title('Kubni spline - 5 zadatih tacaka')
plt.show()
```

```python
# Kubni spline interpolacija
from scipy.interpolate import CubicSpline
cs=CubicSpline(x1,y1)
x_new=np.arange(0,10,0.01)
y_int=cs(x_new)
```

```python
plt.scatter(x1,y1,label='Slucaj 1')
plt.plot(x_new,y_int, label='Interpolirana kriva - 1')
plt.legend()
plt.title('Kubni spline - 5 zadatih tacaka')
plt.show()
```

## Zadatak 2
Posmatrajmo slučaj matematičkog klatna dužine $l$ u homogenom gravitacionom polju. Oscilovanje klatna je opisano diferencijalnom jednačinom:
$$ \dfrac{d^2 \theta}{dt^2}=-\dfrac{g}{l} sin \theta. $$

Za vrlo male uglove $(sin\theta \approx \theta)$ analitičko rješenje daje gdje je rješenje $\theta(t)=\theta_0 cos (\omega t)$, gdje je $\theta_0$ početni ugao za $t=0$, a $\omega=\sqrt(g/l)$. Koristeći Eulerov metod, numerički odrediti vrijednost ugla u funkciji vremena $t$. Posmatrati dva slučaja: kada je $\theta_0=13^{\circ}$ i $\theta_0=50^{\circ}$. U tabelu unijeti odgovarajuće numeričke i analitičke vrijednost i odrediti relativnu diskrepanciju. Nacrtati grafikon koji upoređuje numerička i analitička rješenja za oba slučaja (uzeti da je, T=10 s). 

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

Datu diferencijalnu jednačinu možemo napisati u drugačijem obliku:
$$\dfrac{d\theta} {dt} = \omega$$:
$$ \dfrac{d\omega}{dt}=-\dfrac{g}{l} sin \theta = G (\omega, \theta,t) $$
Koristeći Eulerovu aproksimaciju, rješenje po $\theta$ ćemo naći tako što ćemo paralelno riješavati i po $\omega$ i po $\theta$:
$$ \theta_{n+1}=\theta_n+ h*\omega_n$$
$$ \omega_{n+1}=\omega_n + h* G $$
Potrebno je definisati početne uslove (za $t=0$):
$$\theta_0=13^\circ, \quad \omega_0=0  $$
[$-g/l=-\omega^2=4\pi^2/T^2$]

```python
N = 10000     # number of steps
h = 0.001       # step size

# initial values
t_0 = 0.0
theta_0 = 50*np.pi/180 # radians
w_0 = 0.0

T=10.0
G=4*np.pi**2/T**2

t = np.zeros(N+1)
theta = np.zeros(N+1)
w = np.zeros(N+1)

t[0] = t_0
theta[0] = theta_0
w[0] = w_0

for n in range(N):
    # Euler's method
    theta_new = theta[n] + w[n]*h
    w_new = w[n] - G*np.sin(theta[n])*h
    
    t[n+1] = t[n] + h
    theta[n+1] = theta_new
    w[n+1] = w_new
```

```python
plt.plot(t,theta,label='Numericko rjesenje')
plt.plot(t,theta_0*np.cos(2*np.pi/T*t),':', label='Analiticka aproksimacija')
plt.legend()
plt.xlabel('t(s)')
plt.ylabel('$\theta$ (rad)')
plt.show()
```

```python
#Ispis podataka za tabelu !Vjerovatno ima elegantniji nacin
tt=np.arange(0,10,2)
theta_num=[theta[0],theta[2000],theta[4000],theta[6000],theta[8000]]
theta_an=theta_0*np.cos(2*np.pi/T*tt)
print('t(s),   numericka ,   analiticka ,   relativna greska ')
for i in range(len(tt)):
    print(tt[i],theta_num[i],theta_an[i],(theta_num[i]-theta_an[i])/theta_an[i]*100)
```
